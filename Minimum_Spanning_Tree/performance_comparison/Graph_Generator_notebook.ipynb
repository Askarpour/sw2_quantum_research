```python
import json
```

```python
with open("report_summary.json", "r") as f:
    lista = json.load(f)
```

```python
import matplotlib.pyplot as plt
```

```python
#plt.rc('legend',fontsize=20)
```

```python
w = 20
h = 15
dpi = 300
```

# Fixed shots, time

```python
plt.figure(figsize=(w,h))
for i, s in enumerate([10,100,1000,10000]):
    q = [x for x in lista if x["shots"] == s]
    ax = plt.subplot(2, 2, i+1)
    for name in ["cirq", "qsharp", "qiskit_opt", "qiskit"]:
        x = [(x["qbits"], x[name]["time"]) for x in q]
        x.sort(key=lambda x: x[0])
        y = [z[1] for z in x]
        x = [z[0] for z in x]
        plt.plot(x,y, label=name)
    ax.title.set_text("Time in seconds @ %d shots"%s)
    ax.set_xlabel("qbits")
    ax.set_ylabel("seconds")
    ax.legend()
plt.savefig("imgs/time_fixed_shots_all.png", dpi=dpi)
```
