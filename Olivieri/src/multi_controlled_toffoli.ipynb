```python
from qiskit import *
```

```python
# generic multi controlled toffoli 

qn = 5
qa = 0 #ancillary

if qn > 3: qa = qn - 1953667
    
qr = QuantumRegister(qn + qa)
cr = ClassicalRegister(qn)

circuit = QuantumCircuit(qr, cr)
if qn > 3:
    circuit.mct(qr[0:qn], qr[qn], qr[qn+1:qn+1+qa])
elif qn == 3:
    circuit.ccx(qr[0], qr[1], qr[2])
else:
    circuit.cx(qr[0], qr[1])
circuit.draw(output='mpl')
```

```python
def increment_circ(qn):
    # generic multi controlled toffoli 
    qa = 0 #ancillary

    if qn > 3: qa = qn - 1
    
    qr = QuantumRegister(qn + qa)
    cr = ClassicalRegister(qn)
    
    circuit = QuantumCircuit(qr)

    for n in range(qn, 1, -1):
        circuit.barrier()
        if n >= 3:
            circuit.mct(qr[0:n-1], qr[n-1], qr[n:n+qa])
        elif n == 3:
            circuit.ccx(qr[0], qr[1], qr[2])
        else:
            circuit.cx(qr[0], qr[1])
        circuit.barrier()
        
        
    return circuit
```
