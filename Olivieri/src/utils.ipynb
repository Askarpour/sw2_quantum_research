```python
from qiskit import *
```

```python
circuit = QuantumCircuit(6, 6)
circuit.draw(output='mpl')
```

```python
#put a swap on each qubit and the respective ancillary qubit, for a general n-qubits circuit
def swap_operator(circuit):
    qn_half = int(len(circuit.qubits)/2)
    for q in range(qn_half, 0, -1):
        circuit.swap(qn_half + q -1, q - 1)
    return circuit 
```

```python
swap_operator(circuit)
circuit.draw(output='mpl')
```
